This dataset consists of ~600 songs that were in the top songs of the year from 2010 to 2019 (as measured by Billboard). You can explore interesting song data pulled from [Spotify](http://organizeyourmusic.playlistmachinery.com/#) such as the beats per minute, amount of spoken words, loudness, and energy of every song. 

```python
# Importing all the necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Reading the CSV file
df = pd.read_csv('spotify_top_music.csv', index_col=0)
```

[Source of dataset](https://www.kaggle.com/leonardopena/top-spotify-songs-from-20102019-by-year).

# Objective
### - The objective of this analysis is to find out the most popular Genre of 2010s.
### - This analysis would be useful in forecasting the likelihood of a genre being successful in the future.

```python
# Checking how many unique "top genre" are present in the dataset
len(df['top genre'].unique())
```

```python
# Sorting the dataset according to the popularity
sorted_df = df.sort_values(by = 'pop', ascending=False)
```

```python
# Filtering the dataset where, popularity > 0
removing_0_pop = sorted_df['pop']>0 
sorted_df = sorted_df[removing_0_pop]
```

```python
# New shape of the Dataset is 598. So, 5 rows were eliminated
sorted_df.shape 
```

```python
# Checking the tail of the Dataset if all the tracks with 0 popularity are removed
sorted_df.tail(2)
```

```python
# Checking the head of the Dataset
sorted_df.head(5)
```

```python
# Setting the size of the canvas 
fig_dims = (20, 10) 

# Setting the background color of the canvas
sns.set_style("darkgrid")  

# Setting the font size of the labels
sns.set(font_scale = 2)  

# Unpacking the fig and ax tuple
fig, ax = plt.subplots(figsize=fig_dims)

# Rotating the xticks to 45 degrees
ax.tick_params(axis='x', rotation=90)   

# Setting y-axis label
ax.set(ylabel='No. of Songs')

# Plotting the histogram
sns.histplot(data=sorted_df, x="top genre", ax=ax)
```

# Findings from the Spotify data analysis

#### - From the figure above a trend can be seen that, in last decade **dance pop** was the most popular & dominant _Genre_ among all. 
#### - Therefore it can be said, in the future if a musician makes a track that has the _dance pop Genre_ that is most likely to be a hit.
